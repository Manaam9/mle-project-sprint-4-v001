```python
# Найдем пользоавтелей для тест-кейсов рекомендаций

events = pd.read_parquet("recsys/data/events.parquet")
personal = pd.read_parquet("recsys/recommendations/personal_als.parquet")

print("events columns:", events.columns.tolist())
print("personal columns:", personal.columns.tolist())

print("users in personal:", personal.iloc[:10])
print("users in events:", events.iloc[:10])
```

```python
events = pd.read_parquet("recsys/data/events.parquet")
personal = pd.read_parquet("recsys/recommendations/personal_als.parquet")

personal_users = set(personal["user_id"].unique())
events_users = set(events["user_id"].unique())

offline_and_online = list(personal_users & events_users)[:10]
offline_only = list(personal_users - events_users)[:10]

print("offline_and_online:", offline_and_online)
print("offline_only:", offline_only)
```

```python
# проверяем рекомендации конкретного пользователя

personal = pd.read_parquet("recsys/recommendations/personal_als.parquet")
print(personal.columns.tolist())
print(personal["user_id"].min(), personal["user_id"].max())
print((personal["user_id"] == 999999).any())
```
